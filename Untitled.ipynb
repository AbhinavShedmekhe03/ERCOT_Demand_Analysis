**Importing Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading dataset**

```python
texas2018 = pd.read_csv('texas2018.csv')
texas2019 = pd.read_csv('texas2019.csv')
texas2020 = pd.read_csv('texas2020.csv')
```

**Concatenating Datasets**

```python
df = pd.concat([texas2018, texas2019, texas2020], ignore_index=True)
```

```python
df.head(24)
```

**Cheking for the structure of the dataframe** 

```python
df.info()
```

**Converting appropriate Columns to Numeric**

```python
num_cols = df.columns[1:]
num_cols
```

```python
df[num_cols] = df[num_cols].replace({',': ''}, regex=True).astype(float)
```

```python
df.info()
```

```python
df.describe()
```

The table summarizes the descriptive statistics for electricity demand across different regions in the ERCOT dataset. The 'ERCOT' column represents the total demand. On average, the highest demand is observed in the COAST region, followed by the SOUTH and NCENT regions. The standard deviation indicates the variability in demand, with the highest variability in the SOUTH and COAST regions. The minimum and maximum values provide insights into the range of demand, showcasing the lowest and highest points for each region. The median values (50th percentile) offer a measure of central tendency, indicating the middle point of the distribution. Overall, this summary provides a snapshot of the distribution and characteristics of electricity demand in the ERCOT regions.


**Converting 'HourEnding' to datatime**

```python
# Replace "24:00" with "00:00"
df['HourEnding'] = df['HourEnding'].str.replace('24:00', '00:00')
```

```python
# Remove additional information like " DST" from the 'HourEnding' column
df['HourEnding'] = df['HourEnding'].str.replace(' DST', '')
```

```python
# Convert HourEnding to datetime
df['HourEnding'] = pd.to_datetime(df['HourEnding'], format='%m/%d/%Y %H:%M', errors='coerce')
```

```python
df.info()
```

**Data Visualization**

```python
plt.figure(figsize=(12, 6))
for region in df.columns[1:]:
    plt.plot(df['HourEnding'], df[region], label=region)

plt.xlabel('Hour Ending')
plt.ylabel('Electricity Demand (MWh)')
plt.title('Electricity Demand Over Time')
plt.legend()
plt.show()
```

**Correlation Analysis**

```python
corr_mat = df.iloc[:, 1:].corr()
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
corr_mat
```

- **High Correlations:**

  - **EAST and WEST:** The EAST region shows a strong correlation (0.92) with the WEST region.
  - **NCENT and SCENT:** High correlation (0.94) between the NCENT and SCENT regions.

- **Overall Regional Correlations:**
  - **COAST, EAST, and ERCOT:** The COAST and EAST regions show high correlations (0.89-0.95) with the overall ERCOT demand.

- **Weak Correlations:**
  - **FWEST and NORTH:** The FWEST region has a relatively weak correlation (0.33) with the NORTH region.
  - **WEST and SOUTH:** The WEST region shows a moderate correlation (0.83) with the SOUTH region.

- **General Observation:**
  - **Spatial Dependencies:** The correlations suggest spatial dependencies in electricity demand, indicating that certain regions tend to exhibit similar demand patterns.

**Hourly Analysis**

```python
hourly_demand = df.groupby(df['HourEnding'].dt.hour).mean().iloc[:, 1:]
hourly_demand.plot(kind='bar', stacked=True, figsize=(10,6))
plt.xlabel('Hours of the Day')
plt.ylabel('Average Electricity Demand (MWh)')
plt.title('Average Electricity Demand by Hour of the Day')
plt.show()
```

```python
hourly_demand
```

The hourly electricity demand pattern across ERCOT regions reveals a consistent trend. The COAST region consistently exhibits the highest demand throughout the day, particularly during peak hours from 16:00 to 18:00. The WEST region follows a similar pattern, closely trailing COAST in demand. The NORTH and NCENT regions demonstrate synchronous demand fluctuations, reaching their peaks during the same hours. The EAST and SOUTH regions exhibit relatively lower demand, with minor variations throughout the day. Overall, this pattern highlights the importance of the COAST and WEST regions in contributing to the overall electricity demand in ERCOT, with distinct peak hours observed across the different regions.

**Yearly Comparision**

```python
# Extract the year from 'HourEnding'
df['Year'] = df['HourEnding'].dt.year

# Drop the 'HourEnding' column if it's not needed for further calculations
df = df.drop('HourEnding', axis=1)

# Group by year and calculate the sum for each region
yearly_comparison = df.groupby('Year').sum()

# Plot the results
plt.figure(figsize=(12, 8))
yearly_comparison.T.plot(marker='o')  # Transpose the DataFrame for better visualization
plt.xlabel('Region')
plt.ylabel('Total Demand')
plt.title('Yearly Comparison of Electricity Demand Across Regions')
plt.legend(title='Year', bbox_to_anchor=(1, 1))
plt.show()
```

```python
yearly_comparison
```

- **Yearly Comparison of Electricity Demand Across Regions:**
  - **2018:**
    - The COAST and NCENT regions contributed significantly to the overall electricity demand in 2018, with COAST being the highest.
    - The WEST region had the highest total demand, reaching approximately 376 million units.
  - **2019:**
    - Similar to 2018, COAST and NCENT remained prominent contributors to the total demand, and the WEST region continued to have the highest demand.
    - The overall demand in 2019 slightly increased compared to 2018, reaching around 384 million units.
  - **2020:**
    - The demand dropped significantly in 2020, possibly influenced by various factors such as economic conditions or external events.
    - COAST and NCENT regions maintained their dominance, but the total demand decreased to approximately 112 million units.
  - **General Observations:**
    - COAST, NCENT, and WEST consistently play crucial roles in electricity demand, showcasing the importance of these regions in the ERCOT grid.
    - Monitoring yearly trends helps in understanding the dynamics of electricity consumption and planning for resource allocation and infrastructure development.
